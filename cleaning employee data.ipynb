```python
#Import libraries
import pandas as pd

#Load dataset
df = pd.read_excel("C:/Users/VictoriaAfatsawo/Desktop/Python_cleaning emp_data/employee_survey.xlsx")

df.head()
```

```python
df.info()
```

```python
#Check duplicates 
df.duplicated().sum
```

```python
#Check missing values

df.isnull().sum()
```

```python
#Chech extra spaces
df.apply(lambda col: col.astype(str).str.contains(r'^\s+|\s+$|\s{2,}').sum())
```

```python
#Convert data type
df[["Gender", "MaritalStatus", "JobLevel", "Dept", "EmpType", "CommuteMode", "EduLevel","haveOT"]] = df[["Gender", "MaritalStatus", "JobLevel", "Dept", "EmpType","CommuteMode", "EduLevel", "haveOT" ]].astype("string")
df.dtypes
```

```python
#inconsistent capitalization
df["Gender"] = df["Gender"].str.title()
df["MaritalStatus"] = df["MaritalStatus"].str.title()
df["JobLevel"] = df["JobLevel"].str.title()
df["Dept"] = df["Dept"].str.title()
df["EmpType"] = df["EmpType"].str.title()
df["CommuteMode"] = df["CommuteMode"].str.title()
df["EduLevel"] = df["EduLevel"].str.title()
df["haveOT"] = df["haveOT"].str.title()

#Capitalize 
df["Dept"] = df["Dept"].replace("It", "IT")


```

```python
#Round decimals 
df["PhysicalActivityHours"] = df["PhysicalActivityHours"].round(0).astype("int64")
df["SleepHours"] = df["SleepHours"].round(0).astype("int64")
df["TrainingHoursPerYear"] = df["TrainingHoursPerYear"].round(0).astype("int64")
df 
```

```python
#Exploratory data analysis
#Summary statistic
df.describe()
```

```python
#Save
df.to_excel("cleaned_data.xlsx", index=False)
```
